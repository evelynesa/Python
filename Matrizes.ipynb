```python
import numpy as np
```

```python
matriz01 = np.array([[1,2,3],[4,5,6]])
matriz02 = np.array([[5,5,6],[3,8,4]])

print(matriz01)
print(matriz02)
```

```python
# crie um método que receba duas matrizes, some os valores total de cada matriz e depois multiple o resultado 
#delas e retorne o valor.

def calculaMatriz(mtz01,mtz02):
  print(mtz01.sum() * mtz02.sum())

#Consome o Método
calculaMatriz(matriz01,matriz02)
```

```python
#Exercício 03
#Criar uma matriz nxm (n = 5, m =7(colunas))
matriz03 = np.array([[1,2,3,4,5,6,7],[10,5,9,8,4,3,52],[0,2,5,9,7,1,11],[8,1,6,7,2,6,4],[2,5,8,3,6,9,1]])
#print(matriz03)

#faça a matriz transposta desta matriz
print(matriz03.T)

#somar toda matrix
print(matriz03.sum())

#somar todas as colunas
print(matriz03.sum(axis=0))

#somar todas as linhas.
print(matriz03.sum(axis=1))

#retorne o maior valor
print(matriz03.max())

#retorne o menor valor.
print(matriz03.min())
```
