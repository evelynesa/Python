```python
import pandas as pd
```

```python
#criando dataframes
lista_altura = [1.75,1.73,1.62,1.69,1.62,1.90]
df = pd.DataFrame(lista_altura, columns=['Altura'])
df['Altura']
```

```python
#criando dataframes
lista_altura = [1.75,1.73,1.62,1.69,1.62,1.90]
lista_sexo = ['m','f','f','m','f','m']
df = pd.DataFrame({'Altura':lista_altura, 'Sexo':lista_sexo})
df
```

```python
#Describe Apresenta Dados Quantitativos

#Retorna algumas funções para os dados do dataframe
df.describe()

#retorna todas as opçoes do describe
df.describe(include='all')

#gera uma matriz transposta do dataframe (coluna vira linha e linha vira coluna)
df.describe().transpose
```

```python
#Calcula a média (apenas dos dados quantitativos)
media1 = df.mean()
#calculando médias de colunas específicas (válido tb para mediana, moda ...)
media = df['Altura'].mean()
```

```python
#Calcula a mediana (apenas dos dados quantitativos)
mediana = df.median()
mediana
```

```python
#Calcula a moda (apenas dos dados quantitativos)
moda = df['Altura'].mode()
moda
```

```python
#Calculando os Quantis
quantis = df.quantile([0.1,0.25,0.3,0.5,0.70,0.75,0.99])
quantis
```

```python
#Utilizando List Comprehension em Python

lista_quantis = [x/10 for x in range(1,10)]
quantis = df.quantile(lista_quantis)
quantis
```
