```python
import pandas as pd

df = pd.read_excel('/content/reviews.xlsx')
df = df[df.estrelas == 1]
#df.head()
opinioes = df.texto
opinioes
```

```python
#Frequência do Termo (norm='l2' ou norm='l1')

from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(binary=False, norm='l2', use_idf=False)

#Buscando todas as palavras da página
vetor = vectorizer.fit_transform(opinioes)
pd.DataFrame(vetor.toarray(), columns=vectorizer.get_feature_names())
#print(vectorizer.get_feature_names())
#print(vetor.toarray())
#print(vetor)
```

```python
import nltk, re
nltk.download('rslp')
nltk.download('punkt')
nltk.download('stopwords')
stemmer = nltk.stem.RSLPStemmer()


def stemmer_tokenizer(doc):
  termos = [termo.lower() for termo in nltk.word_tokenize(doc) if re.match(r'^[^\W\d]{3,}$',termo)]
  termos = [stemmer.stem(termo)for termo in termos]
  return termos

stopwords = [stemmer.stem(s) for s in nltk.corpus.stopwords.words('portuguese')]

vectorizer = TfidfVectorizer( stop_words = stopwords, tokenizer = stemmer_tokenizer)

#Buscando todas as palavras da página
vetor = vectorizer.fit_transform(opinioes)
pd.DataFrame(vetor.toarray(), columns=vectorizer.get_feature_names())
#print(vectorizer.get_feature_names())
#print(vetor.toarray())
#print(vetor)
```
