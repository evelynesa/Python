```python
#colocando cada comentário em um arquivo diferente
from bs4 import BeautifulSoup as bs
import requests
import os

def salva_arquivo(pasta, nome, conteudo):  
  if not os.path.isdir(pasta):
    os.makedirs(pasta, exist_ok=True)
  n_files = len(os.listdir(pasta))+1
  with open(f'{pasta}/{nome}_{n_files}.html', 'w', encoding='utf8') as file:
    file.write(conteudo)

def crawler_kindle(base_url,i, n):
  if i <= n:
    url = f'{base_url}{i}'
    response = requests.get(url)
    reviews = bs(response.text, 'html.parser')(attrs={'data-hook':'review'})
    for r in reviews:
      stars = r(class_='a-link-normal')[0].text[0]
      title = r(attrs={'data-hook':'review-title'})[0].text
      text = r(attrs={'data-hook':'review-body'})[0].text
      salva_arquivo(f'/content/Aula02_WebMining/{stars}',f'kindle',f'{title}{text}')
    i+=1
    crawler_kindle(base_url, i, n)
base_url = 'https://www.amazon.com.br/Kindle-preta-sens%C3%ADvel-toque-Gera%C3%A7%C3%A3o/product-reviews/B0186FEYKW/ref=cm_cr_getr_d_paging_btm_next_3?ie=UTF8&reviewerType=all_reviews&pageNumber='
crawler_kindle(base_url,1,10)
```
