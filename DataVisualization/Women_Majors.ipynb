```
import pandas as pd
%matplotlib inline
direct_link = 'http://www.randalolson.com/wp-content/uploads/percent-bachelors-degrees-women-usa.csv'
women_majors = pd.read_csv(direct_link)

print(women_majors.info())
women_majors.head()
```

```
women_majors = women_majors.rename(columns={'Year':'Ano','Art and Performance':'Artes e Performance','Education':'Educação','English':'Inglês','Foreign Languages':'Idiomas Estrangeiros','Health Professions':'Área da Saúde','Public Administration':'Administração Pública'})
women_majors
```

```
# Abaixo de 20%
#under_20 = women_majors.loc[0, women_majors.loc[0]<20]
#under_20

# Acima de 50%
over_50 = women_majors.loc[0, (women_majors.loc[0]>50)&(women_majors.loc[0]<100)]
over_50
```

```
#under_50_graph = women_majors.plot(x = 'Year', y = under_50.index, figsize = (12,8))
#print('Type:', type(under_50_graph))

over_50_graph = women_majors.plot(x = 'Ano', y = over_50.index, figsize = (12,8))
print('Type:', type(over_50_graph))
```
